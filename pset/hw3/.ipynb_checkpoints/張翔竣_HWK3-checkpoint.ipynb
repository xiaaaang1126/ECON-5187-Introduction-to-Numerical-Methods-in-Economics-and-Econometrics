## Draw Data Using Inverse Transform Sampling

A half-normal distribution is often denoted as $N^+(0, \sigma^2)$, which means it is a distribution obtained by truncating a normal distribution $N(0, \sigma^2)$ at the point of `0` and keeping only the positive part (thus the "+" sign) of it. Since it cuts the **underlying normal distribution** into half, it is called a **half-normal distribution**. As you can see from the following graph, the density of $N^+(0, \sigma^2)$ is twice as higher than that of the $N(0, \sigma^2)$.

This homework problem asks you to draw samples from the half-normal distribution $N^+(0, 2)$. It can be done in several ways.

### Define a half-normal distribution using `Distributions.jl`, then use `rand()` to generate 5,000 random values. Draw a histogram of the data and overlay it with the theoretical density plot of $N^+(0,2)$.

 - Hint: You need `Distributions` to define distributions and `StatsPlots` to draw the histogram and the overlaying density plot. You may also need `StatsFuns` to use the normal cdf function.

````julia
# package
using Distributions, StatsPlots


# Q1: (1) Define a half-normal distribution using `Distributions.jl`
#     (2) Use `rand()` to generate 5,000 random values
#     (3) Draw a histogram of the data and overlay it with the theoretical density plot of N^+(0,2)

# (1) define an half-normal distribution with sigma^2 = 2
half_normal_dist = Truncated(Normal(0, sqrt(2)), 0, Inf)

# (2) draw 5,000 samples with rand()
data1 = rand(half_normal_dist, 5000)

# (3) plot both histogram of the data and theoretical density plot
plot1 = histogram(data1, normalize = true, bins = 100, label = "sample Data")
plot1 = plot!(half_normal_dist, label = "theoretical pdf")
println("plot 1: 5,000 samples from rand()")
display(plot1)



# Q2: (1) Define a half-normal distribution using `Distributions.jl`, then use `quantile()` to generate 5,000 random values 
#     (2) Draw a histogram of the data and overlay it with the theoretical density plot of N^+(0,2) 

# (1) draw 5,000 uniform samples as quantile, and transform into half-normla distribution
data2 = quantile(half_normal_dist, rand(5000))

# (2) plot both histogram of the data and theoretical density plot
plot2 = histogram(data2, normalize = true, bins = 100, label = "sample Data")
plot2 = plot!(half_normal_dist, label = "theoretical pdf")
println("plot 2: 5,000 samples from quantile()")
display(plot2)



# Q3: (1) Explain the logic of this line of code. 
#     (2) Draw a histogram of the data and overlay it with the theoretical density plot of N^+(0,2)

# (1) Explain the logic:
#     "Since the given normal distribution is symmetric around 0 and the positive truncated normal is cut at 0 as well,"
#     "the sample drawn from Uni[0.5, 1], which is quantile above 0.5, can be mapped onto the positive truncated normal distribution."
data3 = quantile(Normal(0, sqrt(2)), 0.5*rand(5000) .+ 0.5)

# (2) plot both histogram of the data and theoretical density plot
plot3 = histogram(data3, normalize = true, bins = 100, label = "sample Data")
plot3 = plot!(half_normal_dist, label = "theoretical pdf")
println("plot 3: 5,000 samples from quantile() without truncated function")
display(plot3)



# Q4: Compare the means and standard deviations of the three datasets you generated to the theoretical values
sigma = sqrt(2)
theo_mean = sigma * sqrt(2 / )
theo_var  = sigma^2 * (1 - 2/)
theo_std  = sqrt(theo_var)

mean1, std1 = mean(data1), std(data1)
mean2, std2 = mean(data2), std(data2)
mean3, std3 = mean(data3), std(data3)

println("The theoretical mean is: $theo_mean, and the theoretical standard deviation is: $theo_std")
println("The sample mean in data1 is: $mean1, and the sample standard deviation in data1 is: $std1")
println("The sample mean in data2 is: $mean2, and the sample standard deviation in data2 is: $std2")
println("The sample mean in data3 is: $mean3, and the sample standard deviation in data3 is: $std3")
````

##  In the class, we showed how to transform values from a normal random variable $X \sim F_X(X)$ to values of an exponential random variable $Z \sim F_Z(Z)$. We used a three-step procedure to do this, and we drew graphs to have a visual confirmation on the transformation. This homework problem asks you to modify the graphs we used in class and observe the relationship between the sample size (`n` in the example) and the approximation of the distribution.

### Take a random sample $X$ of `n` measurements from $N(1,2)$ (note: the variance is 2), and draw a histogram of $X$ and compare it to the true $N(1,2)$ density (by overlaying the density to the histogram). Add a slider of $n$ from 100 to 20000. 
- Hint: You need to use packages of `Distributions`, `Plots`, `Interact`, `WebIO`, `StatsPlots`.
- Hint: Use `@manipulate` with a `for` loop.

````julia
# package
using Distributions, Plots, Interact, WebIO, StatsPlots, Statistics

# Draw n samples from N(1, 2) and plot
X = Normal(1, sqrt(2))  
@manipulate for n in (100:20000)

    # draw n samples from N(1, 2)
    random_sample = rand(X, n)

    # plot both histogram of the data and theoretical density plot
    plot = histogram(random_sample, normalize = true, bins = 100, label = "sample data")
    plot = plot!(X, label = "theoretical pdf")
    plot

end
````

### Continue from the above: Transform $X$ to $Y \sim U(0,1)$, and draw a histogram of $Y$ and compare it to a $U(0,1)$. Overlaying the histogram by the density plot of $U(0,1)$.
Hint: You will need `cdf()` function to transform `X` to `Y`.

````julia
# package
using Distributions, Plots, Interact, WebIO, StatsPlots, Statistics

# Draw n samples from N(1, 2) and plot
X = Normal(1, sqrt(2))  
@manipulate for n in (100:20000)

    # draw n samples from N(1, 2)
    random_sample = rand(X, n)

    # plot both histogram of the data from N(1, 2) and theoretical density plot
    plot1 = histogram(random_sample, normalize = true, bins = 100, label = "sample data")
    plot1 = plot!(X, label = "theoretical pdf")
    plot1

    # Consider the cdf of X as random variable, which follows Uni[0, 1]
    Y  = cdf(X, random_sample)

    # plot both histogram of the data from Uni[0, 1] and theoretical density plot
    plot2 = histogram(Y, normalize = true, bins = 100)
    plot2 = plot!(Uniform(0,1), label = "theoretical pdf")

end
````

### Continue from the above: Transform $Y$ to $Z ~ Exp(2)$ (exponential distribution, where mean=standard deviation=2) using an appropriate quantile function, and  draw a histogram of $Z$ and compare it to a Exp$(2)$. Overlay the histogram by a density plot of $Exp(2)$.

Note: If the histogram resembles the density plot, congratulations! You've successfully transformed normally distributed data to exponential distribution data.

````julia
# package
using Distributions, Plots, Interact, WebIO, StatsPlots, Statistics

# Draw n samples from N(1, 2) and plot
X = Normal(1, sqrt(2))  
@manipulate for n in (100:20000)

    # draw n samples from N(1, 2)
    random_sample = rand(X, n)

    # plot both histogram of the data from N(1, 2) and theoretical density plot
    plot1 = histogram(random_sample, normalize = true, bins = 100, label = "sample data")
    plot1 = plot!(X, label = "theoretical pdf")
    plot1

    # Consider the cdf of X as random variable, which follows Uni[0, 1]
    Y  = cdf(X, random_sample)

    # plot both histogram of the data from Uni[0, 1] and theoretical density plot
    plot2 = histogram(Y, normalize = true, bins = 100)
    plot2 = plot!(Uniform(0,1), label = "theoretical pdf")

    # Use Uni[0, 1] as quantile and transform it into exponential
    exp_dist = Exponential(2)
    Z = quantile(exp_dist, Y)
    [mean(Z) std(Z)] |> display

    # plot both histogram of the data from exp(2) and theoretical density plot
    plot3 = histogram(Z, normalize = true, bins = 100)
    plot3 = plot!(exp_dist, label = "theoretical pdf")

end
````

##  The 2nd step in the above transformation relies on PIT (probability integral transformation) which we introduced in the class. However, PIT works best only if the CDF we used in the transformation is close to the true CDF of $X$. This problem asks you to examine what would happen if the CDF used in the transformation is not the actual CDF of $X$.

### Take a random sample $X$ of `n` measurements from $N(1,2)$. Transform $X$ to $Y \sim U(0,1)$ using the CDF of $N(1,1)$. Draw a histogram of $Y$ and overlay it with the density plot of $U(0,1)$. Remember to add a slider of `n`. According to your visual, does $Y$ look like uniformly distributed?

````julia
# package
using Distributions, Plots, Interact, WebIO, StatsPlots, Statistics

# construct two distributions: true normal (X) and falso normal
X = Normal(1, sqrt(2))  
false_normal = Normal(1,1)


@manipulate for n in (100:20000)

    # draw n samples from N(1, 2)
    random_sample = rand(X, n)

    # plot both histogram of the data from N(1, 2) and theoretical density plot
    plot1 = histogram(random_sample, normalize = true, bins=100, label = "sample data")
    plot1 = plot!(X, label = "theoretical pdf")
    plot1

    # use wrong cdf to transform random samples
    Y  = cdf(false_normal, random_sample)

    # plot both histogram of the data given by wrong transforms and theoretical density plot of Uni[0, 1]
    plot2 = histogram(Y, normalize = true, bins = 100, label = "sample data")
    plot2 = plot!(Uniform(0,1), label = "theoretical pdf")

end
````

### Continue from the above: Transform $Y$ to $Z ~ Exp(2)$ (exponential distribution, where mean=standard deviation=2) using an appropriate quantile function, and  draw a histogram of $Z$ and compare it to a Exp$(2)$. Overlay the histogram by a density plot of $Exp(2)$.

Question: Does the histogram match the density plot well? Why not?

````julia
# package
using Distributions, Plots, Interact, WebIO, StatsPlots, Statistics

# construct two distributions: true normal (X) and falso normal
X = Normal(1, sqrt(2))  
false_normal = Normal(1,1)


@manipulate for n in (100:20000)

    # draw n samples from N(1, 2)
    random_sample = rand(X, n)

    # plot both histogram of the data from N(1, 2) and theoretical density plot
    plot1 = histogram(random_sample, normalize = true, bins = 100, label = "sample data")
    plot1 = plot!(X, label = "theoretical pdf")
    plot1

    # use wrong cdf to transform random samples
    Y  = cdf(false_normal, random_sample)

    # plot both histogram of the data given by wrong transforms and theoretical density plot of Uni[0, 1]
    plot2 = histogram(Y, normalize = true, bins = 100, label = "sample data")
    plot2 = plot!(Uniform(0,1), label = "theoretical pdf")

    # Use falso cdf as quantile and transform it into exponential
    exp_dist = Exponential(2)
    Z = quantile(exp_dist, Y)
    [mean(Z) std(Z)] |> display

    # plot both histogram of the data from exp(2) and theoretical density plot
    plot3 = histogram(Z, normalize = true, bins = 100, label = "sample data")
    plot3 = plot!(exp_dist, label = "theoretical pdf")

end
````
